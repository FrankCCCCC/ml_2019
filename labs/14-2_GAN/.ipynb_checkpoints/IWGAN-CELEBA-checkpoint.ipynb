```python
%env CUDA_VISIBLE_DEVICES=0
%matplotlib inline
import os
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow.keras as keras
import imageio
import moviepy.editor as mpy

crt = "/opt/shared-disk/datasets/img_align_celeba_png/"

BATCH_SIZE = 0x100
SAMPLE_COL =  0x4
SAMPLE_ROW = 0x4
SAMPLE_NUM = SAMPLE_COL * SAMPLE_ROW
IMG_H = 0x40
IMG_W = 0x40
IMG_C = 0x3
IMG_SHAPE = (IMG_H, IMG_W, IMG_C)
Z_DIM = 0x100
BUF = 0x10000
IW_LR = 2.0e-04
IW_EPOCH =  0x100
BZ = (BATCH_SIZE, Z_DIM)
BI3 = (BATCH_SIZE, 0x1, 0x1, 0x1)
```

```python
gpus = tf.config.experimental.list_physical_devices('GPU')
tf.config.experimental.set_memory_growth(gpus[0], True)
tf.config.experimental.set_virtual_device_configuration(gpus[0], [tf.config.experimental.VirtualDeviceConfiguration(memory_limit = 0x2400)])
```

```python
# Dataset
AUTOTUNE = tf.data.experimental.AUTOTUNE
iTrain = next(os.walk(crt))[0x2]
dsTrain = tf.data.Dataset.from_tensor_slices(
    iTrain
).map(
    lambda path : tf.image.resize(tf.cast(tf.io.decode_png(tf.io.read_file(crt + path), IMG_C, tf.uint8), tf.float32), (IMG_H, IMG_W)) / 255.0,
    num_parallel_calls = AUTOTUNE
).shuffle(BUF).batch(BATCH_SIZE, True).prefetch(buffer_size = AUTOTUNE)

# Utility function
def utPuzzle(imgs, row, col, path = None):
    h, w, c = imgs[0].shape
    out = np.zeros((h * row, w * col, c), np.uint8)
    for n, img in enumerate(imgs):
        j, i = divmod(n, col)
        out[j * h : (j + 1) * h, i * w : (i + 1) * w, :] = img
    if path is not None : imageio.imwrite(path, out)
    return out
  
def utMakeGif(imgs, fname, duration):
    n = float(len(imgs)) / duration
    clip = mpy.VideoClip(lambda t : imgs[int(n * t)], duration = duration)
    clip.write_gif(fname, fps = n)
```

```python
def GAN(sp, zd):
    # Shape
    xh, xw, xc = sp
    zh = xh // 0x4
    zw = xw // 0x4
        
    # Model
    return keras.Sequential([
        keras.layers.Dense(units  =  0x400, input_shape = (zd,)),
        keras.layers.BatchNormalization(),
        keras.layers.ReLU(),
        keras.layers.Dense(units  =  zh * zw << 0x8),
        keras.layers.BatchNormalization(),
        keras.layers.ReLU(),
        keras.layers.Reshape(target_shape = (zh, zw, 0x100)),
        keras.layers.Conv2DTranspose(
            filters = 0x20,
            kernel_size = 0x5,
            strides = 0x2,
            padding = "SAME"
        ),
        keras.layers.BatchNormalization(),
        keras.layers.ReLU(),
        keras.layers.Conv2DTranspose(
            filters = xc,
            kernel_size = 0x5,
            strides = 0x2,
            padding = "SAME",
            activation = keras.activations.sigmoid
        ),
    ]), keras.Sequential([
        keras.layers.Conv2D(
            filters = 0x20,
            kernel_size = 0x5,
            strides = 0x2,
            padding = "SAME",
            input_shape = sp,
        ),
        keras.layers.LeakyReLU(),
        keras.layers.Conv2D(
            filters = 0x80,
            kernel_size = 0x5,
            strides = 0x2,
            padding = "SAME"
        ),
        keras.layers.BatchNormalization(),
        keras.layers.LeakyReLU(),
        keras.layers.Flatten(),
        keras.layers.Dense(units  =  0x400),
        keras.layers.BatchNormalization(),
        keras.layers.LeakyReLU(),
        keras.layers.Dense(units  =  0x1),
    ])

s = tf.random.normal([SAMPLE_NUM, Z_DIM])
```

```python
IWG, IWD = GAN(IMG_SHAPE, Z_DIM)
opg = keras.optimizers.Adam(IW_LR)
opd = keras.optimizers.Adam(IW_LR)

@tf.function
def IWGTrain(c1):
    z = tf.random.normal([BATCH_SIZE, Z_DIM])

    with tf.GradientTape() as tpg:
        c0 = IWG(z, training = True)

        z1 = IWD(c1, training = True)
        z0 = IWD(c0, training = True)

        ld = tf.reduce_mean(z0)
        lg = - ld
        ld = ld - tf.reduce_mean(z1)
        
        ap = tf.random.uniform(BI3, 0.0, 1.0)
        c2 = (1.0 - ap) * c0 + ap * c1
        with tf.GradientTape() as tpp:
            tpp.watch(c2)
            z2 = IWD(c2)
        gp = tpp.gradient(z2, [c2])[0x0]
        ld += tf.reduce_mean(
            tf.math.squared_difference(
                tf.norm(
                    gp,
                    axis =0x0
                ),
                1.0
            )
        )

    gg = tpg.gradient(lg, IWG.trainable_variables)

    opg.apply_gradients(zip(gg, IWG.trainable_variables))
    
    return lg, ld

@tf.function
def IWDTrain(c1):
    z = tf.random.normal([BATCH_SIZE, Z_DIM])

    with tf.GradientTape() as tpd:
        c0 = IWG(z, training = True)

        z1 = IWD(c1, training = True)
        z0 = IWD(c0, training = True)

        ld = tf.reduce_mean(z0)
        lg = - ld
        ld = ld - tf.reduce_mean(z1)
        
        ap = tf.random.uniform(BI3, 0.0, 1.0)
        c2 = (1.0 - ap) * c0 + ap * c1
        with tf.GradientTape() as tpp:
            tpp.watch(c2)
            z2 = IWD(c2)
        gp = tpp.gradient(z2, [c2])[0x0]
        ld += tf.reduce_mean(
            tf.math.squared_difference(
                tf.norm(
                    gp,
                    axis = 0x0
                ),
                1.0
            )
        )

    gd = tpd.gradient(ld, IWD.trainable_variables)

    opd.apply_gradients(zip(gd, IWD.trainable_variables))
    
    return lg, ld

IWTrain = (
    IWDTrain,
    IWDTrain,
    IWDTrain,
    IWDTrain,
    IWDTrain,
    IWGTrain
)

IWCritic = len(IWTrain)
```

```python
iwlg = [None] * IW_EPOCH
iwld = [None] * IW_EPOCH
iwsp = [None] * IW_EPOCH
rsTrain = float(BATCH_SIZE) / float(len(iTrain))
ctr = 0x0
for ep in range(IW_EPOCH):
    lgt = 0.0
    ldt = 0.0
    for c1 in dsTrain:
        lg, ld = IWTrain[ctr](c1)
        ctr += 0x1
        lgt += lg.numpy()
        ldt += ld.numpy()
        if ctr == IWCritic : ctr = 0x0
    iwlg[ep] = lgt * rsTrain
    iwld[ep] = ldt * rsTrain
    
    out = IWG(s, training = False)
    img = utPuzzle(
        (out * 255.0).numpy().astype(np.uint8),
        SAMPLE_COL,
        SAMPLE_ROW,
        "celebA/iw_%04d.png" % ep
    )
    iwsp[ep] = img
    if ep & 0x1F == 0x1F:
        plt.imshow(img)
        plt.axis("off")
        plt.title("Epoch %d" % ep)
        plt.show()
```

```python
utMakeGif(np.array(iwsp), "celebA/iwgan.gif", duration = 0x2)
```

```python
plt.plot(range(IW_EPOCH), iwld, color = "blue", label = "Discriminator Loss")
plt.plot(range(IW_EPOCH), iwlg, color = "red", label = "Generator Loss")
plt.legend(loc = "upper right")
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Improved WGAN Training Loss")
plt.show()
```
